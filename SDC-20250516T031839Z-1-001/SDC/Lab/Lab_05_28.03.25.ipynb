```python
# Question 6: Write a program to find degree and radian angle between hours and minute hands
# with integers hour and minite
# Step 0: Import library Math
import math
# Step 1: Input hour, minute from keyboard
hour  = int(input("Please input hour  : "))
minute= int(input("Please input minute: "))
# Step 2: Calculate deegree, radian
degree = abs(30*hour - 5.5*minute) if abs(30*hour - 5.5*minute)<180 else (360 - abs(30*hour - 5.5*minute))
#if degree >180:
#  degree = 360-degree
radian = math.radians(degree)
# Step 3: Output
print(f"The angle beetween {hour} hour : {minute} minute")
print(f"\t- Degree : {degree}")
print(f"\t- Radians: {radian:0.2f}")
```

```python
# Question 7: Write a program to Solve the quadratic equation x^2 + 5*x + 6 = 0
# and print step by step (discriminant, check condition of discriminant, solutions)
import math
a,b,c = 1,5,6
delta = b**2 - 4*a*c
if delta > 0:
  print(f"Phương trình có 2 nghiệm phân biệt: ")
  print(f"\t- x1 = {(-b+math.sqrt(delta)/(2*a))}")
  print(f"\t- x2 = {(-b-math.sqrt(delta)/(2*a))}")
elif delta == 0:
  print(f"Phương trình có nghiệm kép: ")
  print(f"x1 = x2 = {-b/(2*a)}")
else:
  print(f"Vô nghiệm")
```

```python
# Question 10: From keyboard input two points with (x,y) and print Euclidean, Manhattan, Cosine
# distance of points
import math
# Step 1: Input A(xA, yA), B(xB, yB)
xA = float(input("xA = "))
yA = float(input("yA = "))
xB = float(input("xB = "))
yB = float(input("yB = "))
# Step 2: Calculate : Euclidean,  Manhattan, Cosin
Euclidean = math.sqrt((xA-xB)**2 +(yA-yB)**2)
Manhattan = abs(xA-xB) + abs(yA-yB)
Cosine    = (xA*xB + yA*yB)/(math.sqrt(xA**2 + yA**2)*math.sqrt(xB**2 + yB**2))
# Step 3: Output
print(f"The distance between A({xA},{yA}) and B({xB},{yB}): ")
print(f"\t- Euclidean: {Euclidean:0.2f}")
print(f"\t- Manhattan: {Manhattan:0.2f}")
print(f"\t- Cosin    : {Cosine:0.2f}")
```

```python
# Question 11: Input from keyboards your birthday with day, month year
# and print information about weekday name, month name, and your age now
from datetime import datetime
dob     = input("Please input your birthday (dd/mm/yyyy): ")
print(type(dob))
dob     = datetime.strptime(dob,"%d/%m/%Y")
print(type(dob))
weekday = dob.strftime("%A")
hour    = dob.strftime("%H")
month   = dob.strftime("%m")
age     = int(datetime.now().strftime("%Y")) - int(dob.strftime("%Y"))
print(f"Your birthday: {dob}")
print(f"\t- Weekday : {weekday}")
print(f"\t- Hour    : {hour}")
print(f"\t- Month   : {month}")
print(f"\t- Age     : {age}")
```

```python
# Question 12: Input four lists of Customer, Product, Quantity
# A. Print out information of customer John
# B. Seperate column QuantityList to Quantity and Unit
# C. Find customer information who bought Pork over 2kg
CustomerList = ["John"     , "John", "Marry"  , "Marry"    , "Marry"]
ProductList  = ["Beer"     , "Pork", "Milk"   , "Vegetable", "Pork" ]
QuantityList = ["2 Bottles", "1 kg", "5 boxes", "2 bunches", "3 kg" ]
# A. Print out information of customer John
indexJohn  = [  x for x in range(0,len(CustomerList)) if CustomerList[x] == "John"]
indexJohn1 = [  x for x,y in enumerate(CustomerList) if y=="John"]
print(indexJohn)
print(indexJohn1)
print(f"The information of John: ")
for x in indexJohn:
  print(f"\t- {ProductList[x]}: {QuantityList[x]}")
# B
Quantity = [ x.split()[0]  for x in QuantityList]
Unit     = [ x.split()[1]  for x in QuantityList]
print(f"Customer: {CustomerList}")
print(f"Product : {ProductList}")
print(f"Quantity: {Quantity}")
print(f"Unit    : {Unit}")
# C. Find customer information who bought Pork over 2kg
indexOver2 = [ x for x,y in enumerate(Quantity) if int(y)>2]
indexKg    = [ x for x,y in enumerate(Unit) if y == "kg"]
indexPork  = [ x for x,y in enumerate(ProductList) if y =="Pork"]
print(f"IndexOver2: {indexOver2}")
print(f"IndexKg   : {indexKg}")
print(f"IndexPork : {indexPork}")
indexSum  = [ x for x in indexKg if ((x in indexOver2) and (x in indexPork))]
print(f"IndexSum  : {indexSum}")
print(f"Customer bought over 2 kg pork: ")
for x in indexSum:
  print(f"\t- {CustomerList[x]}")
```
