```python
#  Các phép toán với kiểu số
#   - Cộng    : +
#   - Trừ     : -
#   - Nhân    : *
#   - Lũy thừa: **
#   - Chia    : /  (chia thập phân)
#   - Chia    : %  (chia lấy dư)
#   - Chia    : // (chia lấy nguyên)
# Ví dụ: Thực  hiện các phép toán số học trên 2 biến a = 10 và b=3
a = 10
b = 3
# Hoán đổi giá trị của a và b
a,b = b,a
print(a)
print(b)
```

```python
a,b=10,3
print(f"{a} + {b} = {a+b}")
print(f"{a} - {b} = {a-b}")
print(f"{a} * {b} = {a*b}")
print(f"{a} ^ {b} = {a**b}")
print(f"{a} / {b} = {a/b:0.3}")
print(f"{a} % {b} = {a%b}")
print(f"{a} // {b} = {a//b}")
```

```python
# Các phép toán boolean (phép so sánh)
# >
# <
# >=
# <=
# ==
# !=
# in
# Ví dụ:
print(3>2)
print(3<2)
print(3<=2)
print(3>=2)
print(3>=3)
print(not(3==3))
```

```python
# Biến 1 thời điểm chỉ lưu được 1 giá trị
# --> Biến 1 thời điểm có thể lưu được nhiều giá trị ?
# --> Biến Collections (Container)
# - list       --> []  (có index --> list[index])
# - tuple      --> ()  (có một phần tử thì phải thêm dấu "," ở cuối)
# - set        --> {}  (không có index, chỉ chứa giá trị không trùng nhau)
# - dictionary --> {key1:value1, key2:value2, key3:value3,...}
# Ví dụ:
DSTen = ["Mr.Khoa", "Mr.Nguyen", "Mr.Dung", "Mr.Chanh", "Mr.Tuan", "Mr.Thong", "Mr.Hieu"]
print(f"Danh sách tên = {DSTen}")
print(f"\t- Kiểu          : {type(DSTen)}")
print(f"\t- Độ dài        : {len(DSTen)}")
print(f"\t- Học viên thứ 1: {DSTen[0]}")
print(f"\t- Học viên thứ 2: {DSTen[1]}")
print(f"\t- Học viên thứ n: ....")
print(f"\t- Học viên thứ 7: {DSTen[6]}")
DSTuoi = {43, 43, 24, 27, 20, 20, 20}
print(f"Tuổi: {DSTuoi}")
print(f"\t- Kiểu          : {type(DSTuoi)}")
print(f"\t- Độ dài        : {len(DSTuoi)}")
gioiTinh = ("Nam",)
print(f"Giới tính = {gioiTinh}")
print(f"\t- Kiểu          : {type(gioiTinh)}")
print(f"\t- Độ dài        : {len(gioiTinh)}")
```
