```python
import random
# function sortBubble
def sortBubble(arr):
  for last in range(len(arr)-1,0,-1):
    for inter in range(last):
      if arr[inter]>arr[inter+1]:
        arr[inter],arr[inter+1] = arr[inter+1], arr[inter]
  return arr
n = int(input("Enter number of array: "))
unsortArr = []
for x in range(n):
  unsortArr.append(random.randint(1,10))
print(f"Unsorted array: {unsortArr}")
sortedArr = sortBubble(unsortArr)
print(f"Sorted array: {sortedArr}")
```

```python
class BubbleSort:
  def __init__(self, size = 5):
    self.__nItems = 0
    self.__a = [None]*size
  @property
  def nItems(self):
    return self.__nItems
  @nItems.setter
  def nItems(self, nItem):
    self.__nItems = nItem
  @property
  def a(self):
    return self.__a
  @a.setter
  def a(self,a):
    self.__a=a
  def fill(self):
    for x in range(len(self.__a)):
      self.__a[x]=random.randint(1,10)
      self.__nItems+=1
  def sortBubble(self):
    for last in range(len(self.__a)-1,0,-1):
      for inter in range(last):
        if self.__a[inter]>self.__a[inter+1]:
          self.__a[inter],self.__a[inter+1] = self.__a[inter+1], self.__a[inter]
    return self.__a
  def run(self):
    self.fill()
    print(f"Unsorted array: {self}")
    print(f"Sorted array: {self.sortBubble()}")
  def __str__(self):
    st=""
    st+=f"{self.__a}"
    return st

n = int(input("Enter number of array: "))
b = BubbleSort(n)
b.run()

```
