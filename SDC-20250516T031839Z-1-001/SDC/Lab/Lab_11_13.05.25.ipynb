```python
from abc import ABC, abstractclassmethod, abstractmethod
class Sach (ABC):
  # Thuộc tính: MaSach, TieuDe, TacGia, NamXuatBan
  # Hàm dựng
  def __init__(self, MaSach, TieuDe, TacGia, NamXuatBan):
    self.__MaSach     = MaSach
    self.__TieuDe     = TieuDe
    self.__TacGia     = TacGia
    self.__NamXuatBan = NamXuatBan
  # Getters/setters
  @property
  def MaSach(self):
    return self.__MaSach
  @MaSach.setter
  def MaSach(self, MaSach):
    self.__MaSach = MaSach
  @property
  def TieuDe(self):
    return self.__TieuDe
  @TieuDe.setter
  def TieuDe(self, TieuDe):
    self.__TieuDe = TieuDe
  @property
  def TacGia(self):
    return self.__TacGia
  @TacGia.setter
  def TacGia(self, TacGia):
    self.__TacGia = TacGia
  @property
  def NamXuatBan(self):
    return self.__NamXuatBan
  @NamXuatBan.setter
  def NamXuatBan(self, NamXuatBan):
    self.__NamXuatBan = NamXuatBan
  # Phương thức
  @abstractmethod
  def thanhTien(self):
    pass
  def __str__(self):
    st=""
    st+=f"{self.__MaSach:5} | {self.__TieuDe:20} | {self.__TacGia:15} | {self.__NamXuatBan:5} | "
    return st
```

```python
class SachGiaoKhoa(Sach):
  # Thuộc tính: SoTrang, PhuThu
  # Hàm dựng
  def __init__(self,MaSach, TieuDe, TacGia, NamXuatBan, SoTrang, PhuThu):
    super().__init__(MaSach, TieuDe, TacGia, NamXuatBan)
    self.__SoTrang = SoTrang
    self.__PhuThu  = PhuThu
  # Getters/Setters
  @property
  def SoTrang(self):
    return self.__SoTrang
  @SoTrang.setter
  def SoTrang(self, SoTrang):
    self.__SoTrang = SoTrang
  @property
  def PhuThu(self):
    return self.__PhuThu
  @PhuThu.setter
  def PhuThu(self, PhuThu):
    self.__PhuThu = PhuThu
  # Phương thức
  def thanhTien(self):
    return self.__SoTrang*200 + self.__PhuThu
  def __str__(self):
    st=super().__str__()
    st+=f"{self.__SoTrang: 5} | {self.__PhuThu:10} |"
    return st
```

```python
class SachTieuThuyet(Sach):
  # Thuộc tính: SoTrang, TheLoai, Rating
  # Hàm dựng
  def __init__(self, MaSach, TieuDe, TacGia, NamXuatBan, SoTrang, TheLoai, Rating):
    super().__init__(MaSach, TieuDe, TacGia, NamXuatBan)
    self.__SoTrang = SoTrang
    self.__TheLoai = TheLoai
    self.__Rating  = Rating
  @property
  def SoTrang(self):
    return self.__SoTrang
  @SoTrang.setter
  def SoTrang(self, SoTrang):
    self.__SoTrang = SoTrang
  @property
  def TheLoai(self):
    return self.__TheLoai
  @TheLoai.setter
  def TheLoai(self, TheLoai):
    self.__TheLoai = TheLoai
  @property
  def Rating(self):
    return self.__Rating
  @Rating.setter
  def Rating(self, Rating):
    self.__Rating = Rating
  # Phương thức
  def thanhTien(self):
    return self.__SoTrang*500 + 0.01*self.__Rating*self.__SoTrang*500
  def __str__(self):
    st= super().__str__()
    st+=f"{self.__SoTrang: 5} |            |{self.__TheLoai:10}| {self.__Rating:5}"
    return st
```

```python
class QuanLySach:
  # Thuộc tính:
  # Hàm dựng
  def __init__(self):
    self.__DanhSach = []
  @property
  def DanhSach(self):
    return self.__DanhSach
  @DanhSach.setter
  def DanhSach(self, DanhSach):
    self.__DanhSach = DanhSach
  # Phương thức
  def themSach(self, s):
    self.__DanhSach.append(s)
  def hienThiSach(self):
    print(f"Mã sách | Tiêu đề                 | Tác giả        | NXB | Số trang | Thể Loại | Rating ")
    for s in self.__DanhSach:
      print(s)
  def tongGiaSach(self):
    sum=0
    for s in self.__DanhSach:
      sum+=s.thanhTien()
    print(f"TỔNG TIỀN : {sum} VNĐ")
```

```python
sgk1 = SachGiaoKhoa("001", "Văn","Nguyen Van A","2024",200,2000)
stt1 = SachTieuThuyet("001", "Văn","Nguyen Van A","2024",200,"Trinh Tham",5)
ql   = QuanLySach()
ql.themSach(sgk1)
ql.themSach(stt1)
ql.hienThiSach()
ql.tongGiaSach()
```
