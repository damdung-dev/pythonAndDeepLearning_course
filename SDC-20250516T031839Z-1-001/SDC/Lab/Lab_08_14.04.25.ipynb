```python
class Student:
  # Attributes
  count = 0
  # Constructor
  def __init__(self, name, age):
    self.__name = name
    self.__age  = age
    Student.count+=1
    self.__id   = Student.count
  # getters, setters
  @property
  def name(self):
    return self.__name
  @name.setter
  def name(self, name):
    self.__name = name
  @property
  def age(self):
    return self.__age
  @age.setter
  def age(self, age):
    self.__age = age
  @property
  def id(self):
    return self.__id
  @id.setter
  def id(self, id):
    self.__id = id
  # Methods
  def greeting(self):
    print(f"=====Hello=====")
  # __str__()
  def __str__(self):
    st=f"My name is {self.__name}"
    st+=f"\nI am {self.__age} years old"
    st+=f"\nMy ID is {self.__id}"
    return st
  def introduce(self):
    print(f"My name is {self.__name}")
    print(f"I am {self.__age} years old")
    print(f"My ID is {self.__id}")
s1 = Student("Chanh", 24)
s2 = Student("Dung", 22)
s1.greeting()
print(s1)
s2.greeting()
print(s2)
```
