```python
# Biến - nơi lưu trữ thông tin
# Khai báo biến: tên_biến = giá1_trị ;
# Kiểu dữ diệu:
#    - Kiểu chuỗi (string  --> str)
#    - Kiểu số:
#        + Kiểu số nguyên (interger - int)
#        + Kiểu số thực   (float, double)
#    - Kiểu boolean
```

```python
name   = "Thìn"
age    = 38
city   = "Da Nang"
height = 165.0
print("Xin chào")
print("Tôi tên là ",name)
print("Tôi năm nay",age, "tuổi")
print("Hiện đang ở tại", city)
print("Tôi cao",height,"cm")
print("Kiểm tra kiểu dữ liệu")
print(type(name))
print(type(age))
print(type(city))
print(type(height))
```

```python
# Lệnh nhập/ xuất trong python (I/O)
# 2. Lệnh nhập: input()
name   = input("Mời bạn nhập tên      : ")
age    = int(input("Mời bạn nhập tuổi     : "))
city   = input("Mời bạn nhập nơi ở    : ")
height = float(input("Mời bạn nhập chiều cao: "))
print("="*40)
print(f"Thông tin của bạn là: ")
print(f"\t- Tên      : {name}")
print(f"\t- Tuổi     : {age}")
print(f"\t- Nơi ở    : {city}")
print(f"\t- Chiều cao: {height}")
if (age>60):
  print("Gia")
```
