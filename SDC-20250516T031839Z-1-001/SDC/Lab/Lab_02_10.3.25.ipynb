```python
# Xử lý chuỗi
# Indexing (slicing)
# Syntax: [ begin :  end   : step ]
#         [include: exclude: >/<  ]
# Hàm len() --> trả về độ dài chuỗi
s = "Hello Viet Nam"
l = len(s)
print(l)
print(s[13])
print(s[len(s)-1])
# Hello
print(s[0:5:1])
print(s[:5:])
print(s[:5])
# Nam
print(s[11:14:1])
print(s[-3::1])
print(s[11:14])
print(s[11:])
# Viet
s1 = "Hello Viet Nam"
print(s[6:10:1])
print(s[-8:-4:1])
# loVe
print(s[2:9:2])
```
