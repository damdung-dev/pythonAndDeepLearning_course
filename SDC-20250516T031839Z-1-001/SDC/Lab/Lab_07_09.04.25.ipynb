**Bài 1.**	Viết một chương trình tính giai thừa của một số cho trước. Kết quả được in thành chuỗi trên một dòng. Ví dụ, số cho trước là 8 thì kết quả đầu ra phải là 40320.

```python
# Định nghĩa hàm
def tinhGiaiThua(n):
  kq = 1
  for x in range(1,n+1):
    # kq = kq * x
    kq*=x
  print(f"Giai thừa của {n} là {kq}")
def tinhGiaiThua1(n):
  kq=1
  for x in range(1,n+1):
    kq*=x
  return kq
a = tinhGiaiThua(8)
b = tinhGiaiThua1(8)
print(a)
print(b)
def tinhGiaiThua3(n):
  if (n==0 or n==1):
    return 1
  else:
    return n* tinhGiaiThua3(n-1)
```

**Bài 2.**	Định nghĩa một hàm có đầu vào là 2 chuỗi và in chuỗi có độ dài lớn hơn ra màn hình. Nếu 2 chuỗi có chiều dài như nhau thì in tất cả các chuỗi theo từng dòng. Sử dụng hàm len() để lấy chiều dài của một chuỗi.

```python
def inChuoi(s1,s2):
  if(len(s1)>len(s2)):
    print(s1)
  else:
    if (len(s1)<len(s2)):
      print(s2)
    else:
      print(s1)
      print(s2)
```

```python
def inChuoi1(s1,s2):
  if (len(s1)>len(s2)):
    print(s1)
  elif (len(s1)<len(s2)):
    print(s2)
  else:
    print(s1)
    print(s2)
```

```python
inChuoi("Chanh","Dung")
inChuoi1("Chanh", "Dung")
```

**Bài 3.**	Định nghĩa một hàm có thể tạo và in một danh sách chứa các giá trị bình phương của các số từ 1 đến 20 (tính cả 1 và 20). Yêu cầu: Sử dụng toán tử ** để lấy giá trị bình phương; Sử dụng range() cho vòng lặp; Sử dụng list.append() để thêm giá trị vào danh sách.

```python
def inBinhPhuong():
  list1=[]
  for x in range(1,21):
    a=x**2
    list1.append(a)
  return list1
def inBinhPhuong1():
  return [x**2 for x in range(1,21)]
print(inBinhPhuong())
print(inBinhPhuong1())
```

**Bài 9.**	Số nguyên tố là số nguyên lớn hơn 1 và chỉ chia hết cho một và chính nó. Viết hàm có 1 tham số để xác định tham số đó có phải là số nguyên tố hay không, trả về True nếu là số nguyên tố và ngược lại trả về False. Viết chương trình chính đọc số nguyên từ người dùng và hiển thị thông báo cho biết đó có phải là số nguyên tố hay không.



```python
def kiemTraSNT():
  number = int(input("Mời bạn nhập số nguyên bất kỳ: "))
  kt     = True
  for x in range(2,number):
    if number%x==0:
      kt = False
  if kt:
    print(f"{number} là số nguyên tố")
  else:
    print(f"{number} không phải là số nguyên tố")
kiemTraSNT()
```

```python
# In ra số nguyên tố từ 1 đến 100
def inSNT():
  for x in range(2,101):
    kt = True
    for y in range(2,x):
      if (x%y==0):
        kt = False
        break
    if kt:
      print(x, end=", ")
inSNT()
```

```python
# In ra các số không phải số nguyên tố từ 1 đến 100
def inKSNT():
  for x in range(1,101):
    if x==1:
      print(x, end=", ")
      continue
    kt = True
    for y in range(2,x):
      if x%y ==0:
        kt = False
        break
    if kt:
      continue
    else:
      print(x, end=", ")
inKSNT()
```
