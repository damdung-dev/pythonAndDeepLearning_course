**Bài 1.**	Hai từ là đảo chữ cái nếu chúng chứa tất cả các chữ cái giống nhau, nhưng theo một thứ tự khác nhau. Ví dụ, “evil” và “live”  là một đảo chữ vì từ đều chứa “e”, “i”, “l”, và “v”. Tạo một chương trình đọc hai chuỗi từ người dùng, xác định xem chúng có phải là đảo chữ hay không và báo cáo kết quả.

```python
def checkReverse(a, b):
  return (b[::-1] == a)

a = "evia"
b = "live"
print(checkReverse(a, b))
```
